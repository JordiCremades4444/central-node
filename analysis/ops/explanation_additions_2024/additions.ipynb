# Additions

## Executive summary

An addition occurs whan a non-originally placed item is added to an order. However, this product should not be the replacement of another. A common example of additions are plastic bags that need to be added to an order. In data terms, additions create 2 registres in bought products. A first register with the product with quantity zero, which is replaced by a product with the same external id and quantity greater than zero.

During the last 30 days, in Groceries partners, around 20% of orders contain additions.


**Business questions answered:**

- **How do additions differentiate to price mismatch?** --> Additions create 2 registers in bought products table. The first one with quantity 0. The second one with quantity >0. However, the price mismatch also has to registers in bought products table. The first one with quantity>0, and the second register with quantity <> quantity of the first register

- **How do additions differentiate to replacements?** --> Additions create 2 registers with the same external id. Replcaments have 2 registers with different external id^

- **How do price mismatch and partial removal differentiate?** --> There is no way to distinguish. Both create 2 registers, and change the quantity of the products. Obviously, in both scenarios the final price of the product will have change. Thus, we cannot actively differentiate between price mismatch and partial removal
